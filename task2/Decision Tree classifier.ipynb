# importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
```

# Load Dataset

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# visulization

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')
```

```python
sns.pairplot(df,hue='species')
```

```python
# selecting target variable
x=df.drop('species',axis=1)
y=df['species']
```

# split data

```python
# splittig data into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# fit and train model

```python
# fitting model with DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
# Predicting the valuse
ypred=model.predict(x_test)
```

# Evaluation of model

```python
# evaluation of model
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(y_test,ypred)
print("accuray Score :",accuracy_score(y_test,ypred))
print(classification_report(y_test,ypred))
sns.heatmap(cm,annot=True)
```

# Graphically Representation of Decision Tree

```python
features=x.columns
label=y.unique()
print(label)

plt.figure(figsize=(15,15))
tree.plot_tree(model,filled=True,class_names=label,feature_names=features)
```
